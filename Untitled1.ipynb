```python
data = xr.open_dataset('/Users/dlee/Downloads/tmax.2020.nc')
data
```

```python
dateDay = '2020-09-01'
fn_out = '/Users/dlee/Downloads/tmax.2020.09.01.nc'
data_africa_east = data.sel(time=slice(dateDay, dateDay), lat=slice(20,-15), lon=slice(28, 54))
# Reset actual range
lat_actual = np.array([data_africa_east.lat.max(), data_africa_east.lat.min()])
data_africa_east.lat.attrs.update({'actual_range':lat_actual})
lon_actual = np.array([data_africa_east.lon.min(), data_africa_east.lon.max()])
data_africa_east.lon.attrs.update({'actual_range':lon_actual})
data_africa_east.to_netcdf(fn_out)
print('%s is saved..' % fn_out)
```
