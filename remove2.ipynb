```python
# Load Dam Inflow data from SUTD
dfFlowDams = pd.read_hdf('./data/dfFlowDams.hdf')
ind_dams = np.load('./data/ind_dams.npz')['ind_dams']
# Load Degree of Regulation (DOR)
dor = pd.read_hdf('./data/new_dor.hdf')      # The order is sorted during the process
dam_dor = dor.loc[dor.DOR2 <= 0, 'GRAND_ID']
# Select only headwater dams (735)
# idx_dor = np.isin(ind_dams[0,:], dam_dor)
# damList = ind_dams[0,idx_dor]
# ind_dams = ind_dams[1,idx_dor]
damList = ind_dams[0]
ind_dams = ind_dams[1]
ndam = len(damList)
```

```python

mp3 = pd.read_hdf('./data/prediction_3mp/dfMP1.hdf')
mp5 = pd.read_hdf('./data/prediction_5mp/dfMP1.hdf')
mp7 = pd.read_hdf('./data/prediction/dfMP1.hdf')

comp = np.zeros([ndam,3])
corr = comp.copy()
for i,did in enumerate(damList):
    comp[i,0] = he.kge_2012(dfFlowDams[did], mp3[did])
    comp[i,1] = he.kge_2012(dfFlowDams[did], mp5[did])
    comp[i,2] = he.kge_2012(dfFlowDams[did], mp7[did])
    corr[i,0] = stats.pearsonr(dfFlowDams[did], mp3[did])[0]
    corr[i,1] = stats.pearsonr(dfFlowDams[did], mp5[did])[0]
    corr[i,2] = stats.pearsonr(dfFlowDams[did], mp7[did])[0]  
    
comp = pd.DataFrame(index = damList, columns=['kge_mp3', 'kge_mp5', 'kge_mp7'], data=comp).T
corr = pd.DataFrame(index = damList, columns=['corr_mp3', 'corr_mp5', 'corr_mp7'], data=corr).T
```
